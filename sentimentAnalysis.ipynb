```python
from transformers import pipeline
model = pipeline('sentiment-analysis')
```

```python
model('i am happy')
```

```python
model("i killed a women's baby")
```

```python
res = model(['it is a bad day', 'i will fight for my life', 'it is nice'])
for i in res:
  print(f"label: {i['label']} and score: {i['score']}")
```

```python
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification
```

```python
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"
modell = TFAutoModelForSequenceClassification.from_pretrained(model_name, from_pt=True)
tokenizer = AutoTokenizer.from_pretrained(model_name)
m = pipeline('sentiment-analysis', model=modell, tokenizer=tokenizer)
```

```python
m('i am bad boy')
```
